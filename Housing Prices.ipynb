```python
#Droping rows with null
housing_data_numeric = housing_data_numeric.dropna()
```

```python
from sklearn.preprocessing import StandardScaler

X = housing_data_numeric.drop(['SalePrice'], axis=1)
y = housing_data_numeric['SalePrice']

scaler = StandardScaler()
X_data = scaler.fit_transform(X)
y_data = y.to_numpy()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=1)

# Print the shapes of the resulting sets (optional)
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error as mse
from math import sqrt

# Step 2: Train the model
lr = LinearRegression()
lr.fit(X_train, y_train)

# Step 3: Make predictions on the test set
y_pred = lr.predict(X_test)

# Step 4: Calculate RMSE
rmse = sqrt(mean_squared_error(y_test, y_pred))
print(f'Root Mean Squared Error (RMSE): {rmse}')
```

```python
from sklearn.metrics import r2_score


r2 = r2_score(y_test, y_pred)
print(f'R-squared Score: {r2}')
```

```python
lm = LinearRegression()
lm = lm.fit(X_train, y_train)
mse(lm.predict(X_train), y_train, squared=False)
```
